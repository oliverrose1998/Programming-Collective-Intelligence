Cluster the blog data using the Hierachical Clustering and draw the dendrogram to show clustering of blogs. Rotate the data and repeat to produce a dendrogram to show clustering of words.

```python
blognames, words, data = clusters.readfile('blogdata.txt')
```

```python
clust = clusters.hcluster(data)
clusters.drawdendrogram(clust,blognames,jpeg='blogclust.jpg')

rdata = clusters.rotatematrix(data)
wordclust = clusters.hcluster(rdata)
clusters.drawdendrogram(wordclust,words,jpeg='wordclust.jpg')
```

Cluster blog data using K-Means Clustering

```python
kclust = clusters.kcluster(data, k=10)

for r in kclust[0]:
    print(blognames[r]) 
```

Use Multidimensional Scaling to produce a 2D representation of the similarity of the data. Similar blogs will be represented closer together.

```python
blognames, words, data = clusters.readfile('blogdata.txt')
coords = clusters.scaledown(data)
clusters.draw2d(coords,blognames,jpeg='blogs2d.jpg')
```
